```python
import numpy as np
def sumLargestNumbers(array):
    # Put your solution here
    array = np.sort(array)
    new_array = array[-2:]
    sum = np.sum(new_array)
    return sum

print(sumLargestNumbers([1, 10]))
print(sumLargestNumbers([1, 2, 3]))
print(sumLargestNumbers([10, 4, 34, 6, 92, 2]))
```

```python
def conditionalSum(values, condition):
    # Your code here
    sum = 0
    value = 0
    while value < len(values):
        if values[value] % 2 == 0 and condition == "even":
            sum = sum + values[value]
            value += 1
        elif values[value] % 2 == 1 and condition == "odd":
            sum = sum + values[value]
            value += 1
        else:
            value += 1
    return sum
            
            


print(conditionalSum([1, 2, 3, 4, 5], "even"))
print(conditionalSum([1, 2, 3, 4, 5], "odd"))
print(conditionalSum([13, 88, 12, 44, 99], "even"))
print(conditionalSum([], "odd"))
```

```python
def numberOfVowels(data):
    # Put your solution here
    #a,e,i,o,u
    vowel_count = 0
    letter = 0
    while letter < len(data):
        if data[letter] == 'a':
            vowel_count += 1
            letter += 1
        elif data[letter] == 'e':
            vowel_count += 1
            letter += 1
        elif data[letter] == 'i':
            vowel_count += 1
            letter += 1
        elif data[letter] == 'o':
            vowel_count += 1
            letter += 1
        elif data[letter] == 'u':
            vowel_count += 1
            letter += 1
        else:
            letter += 1
    return vowel_count
            
#I know there's a quicker way to do this without 6 if statements but i can't figure out the syntax :/.

print(numberOfVowels("orange"))
print(numberOfVowels("lighthouse labs"))
print(numberOfVowels("aeiou"))
```

```python
def instructorWithLongestName(instructors):
    # Put your solution here
    current_max = 0
    return_key = {}
    for key in instructors:
        name = key.get("name")
        current = len(name)
        if current_max < current:
            current_max = current
            return_key = key
            
    return return_key
        


print(instructorWithLongestName([
  {"name": "Samuel", "course": "iOS"},
  {"name": "Jeremiah", "course": "Data"},
  {"name": "Ophilia", "course": "Web"},
  {"name": "Donald", "course": "Web"}
]))
print(instructorWithLongestName([
  {"name": "Matthew", "course": "Data"},
  {"name": "David", "course": "iOS"},
  {"name": "Domascus", "course": "Web"}
]))
```

```python
def urlEncode(text):
    # Put your solution here
    encoded_text = []
    for letter in text.strip():
        if letter != ' ':
            encoded_text = np.append(encoded_text, letter)
        elif letter == ' ':
            encoded_text = np.append(encoded_text, "%20")
            
    return "".join(encoded_text)

print(urlEncode("Lighthouse Labs"))
print(urlEncode(" Lighthouse Labs  "))
print(urlEncode("blue is greener than purple for sure"))
```

```python
def whereCanIPark(spots, vehicle):
    # Code here!
    coordinates = []
    
    #loop that keeps track of value and index of each row
    for count_y, y in enumerate(spots):
        
        #loop that keeps track of value and index of each column
        for count_x, x in enumerate(y):
            
            if vehicle == 'regular' and x == 'R':
                coordinates.append(count_x)
                coordinates.append(count_y)
                return coordinates
            
            elif vehicle == 'small' and x in ('R','S'):
                coordinates.append(count_x)
                coordinates.append(count_y)
                return coordinates
            
            elif vehicle == 'motorcycle' and x in ('R','S','M'):
                coordinates.append(count_x)
                coordinates.append(count_y)
                return coordinates 
        
    return False

print(whereCanIPark(
  [
    # COLUMNS ARE X
    # 0    1    2    3    4    5
    ['s', 's', 's', 'S', 'R', 'M'], # 0 ROWS ARE Y
    ['s', 'M', 's', 'S', 'r', 'M'], # 1
    ['s', 'M', 's', 'S', 'r', 'm'], # 2
    ['S', 'r', 's', 'm', 'r', 'M'], # 3
    ['S', 'r', 's', 'm', 'r', 'M'], # 4
    ['S', 'r', 'S', 'M', 'M', 'S']  # 5
  ],
  'regular'
))

print(whereCanIPark(
  [
    ['M', 'M', 'M', 'M'],
    ['M', 's', 'M', 'M'],
    ['M', 'M', 'M', 'M'],
    ['M', 'M', 'r', 'M']
  ],
  'small'
))

print(whereCanIPark(
  [
    ['s', 's', 's', 's', 's', 's'],
    ['s', 'm', 's', 'S', 'r', 's'],
    ['s', 'm', 's', 'S', 'r', 's'],
    ['S', 'r', 's', 'm', 'r', 's'],
    ['S', 'r', 's', 'm', 'R', 's'],
    ['S', 'r', 'S', 'M', 'm', 'S']
  ],
  'motorcycle'
))
```

```python
def checkAir(samples, threshold):
    # Code here!
    dirty_samples = 0
    #counts dirty samples and total
    for count,sample in enumerate(samples):
        if sample == 'dirty':
            dirty_samples += 1
    
    #accounting for the fact counts first value was 0 rather than 1
    count += 1
    
    #calculating the percentage of dirty samples to the total
    percentage = dirty_samples/count
    
    #evaluating if the percentage is over or under the threshold
    print(percentage)
    if percentage < threshold:
        return 'Clean'
    elif percentage > threshold:
        return 'Polluted'
        

print(checkAir(
  ['clean', 'clean', 'dirty', 'clean', 'dirty', 'clean', 'clean', 'dirty', 'clean', 'dirty'],
  0.3
))

print(checkAir(
  ['dirty', 'dirty', 'dirty', 'dirty', 'clean'],
  0.25
))

print(checkAir(
  ['clean', 'dirty', 'clean', 'dirty', 'clean', 'dirty', 'clean'],
  0.9
))
```
